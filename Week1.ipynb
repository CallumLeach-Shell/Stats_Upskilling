## Probability Density Function
A probability density function (PMf when $X$ is a discrete RV), is a function whose value at any given sample in the sample space can be interpreted as providing a relative likelihood that the value of the random variable would be equal to that sample.
A random variable $X$ has density $f_{X}$, where

$Pr[a \leq X \leq b] = \int_{a}^{b}{f_{X}(x)}\,dx$

Hence, if $F_X$ is the cumulative fistribution function of $X$, then:

$F_{X}(x) = \int_{-\infty}^{x}{f_{X}(u)}\,du$

$f_X(x) = \frac{d}{dx}F_X(x)$

Which represents the probability that $X$ will take a value less than or equal to $x$

```python
import scipy.stats as stats
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# Initialise normal class, with mean 100 and s.d 15
normal = stats.norm(loc=100, scale=15)

# This code will calculate the pdf or the points within the linspace.
t = np.linspace(50,150,100)
plt.plot(t, normal.pdf(t), label="PDF")
plt.legend()
plt.show()
```

- The area under the curve is the probability, hence the integral above.
- The total area under the curve must equal 1 (i.e $\lim_{x \to \infty}{F_X(x)} = 1$)
